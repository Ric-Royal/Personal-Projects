## Extracting images

```python
# Loading libraries to help in converting the pixel data to images. While using the last column as the label

import numpy as np
import csv
from PIL import Image

counter = dict()

with open('locust.csv') as csv_file:
  csv_reader = csv.reader(csv_file)

  # skip the header row
  next(csv_reader)

  for row in csv.reader(csv_file):

    pixels = row[:-1] # exclude the label column
    pixels = np.array(pixels,dtype= 'float64')
    pixels = pixels.reshape((138, 138, 3))
    image = Image.fromarray((pixels*255).astype(np.uint8))

    label = row[-1]

    if label not in counter:
      counter[label] = 0
    counter[label] += 1

    filename = '{}{}.jpg'.format(label, counter[label])
    image.save(filename)

    print('saved:', filename)

```
